```python
import pandas as pd
from pandas import Series, DataFrame
```

```python
from pandas import read_html
```

```python
d1 = pd.io.html.read_html('http://www.lotto-8.com/listlto539.asp?indexpage=1&orderby=new') [4]

type(d1)
d1
```

```python
for i in range(2,5):
    url = 'http://www.lotto-8.com/listlto539.asp?indexpage='+ str(i) +'&orderby=new'
    d2 = pd.io.html.read_html(url)[4][1:]
    d1 = d1.append(d2,ignore_index = True)
d1
```

```python
writer = pd.ExcelWriter('output.xlsx')
d1.to_excel(writer,sheet_name='Sheet1', na_rep='')
writer.save()
```
